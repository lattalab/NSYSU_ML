```python
# all library will used
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.linear_model import Ridge, Lasso
```

#### read the csv file  
#### 對日期欄位做處理: 轉成datetime object再轉成float  
#### 這樣一來, 日期也能當作feature的input

#### 還有填補缺失值

```python
# translate to datetime object in python
df = pd.read_csv('HW2_bike-sharing_train.csv')
df['datetime'] = pd.to_datetime(df['datetime'])
df['datetime'] = df['datetime'].values.astype(float).reshape(-1, 1)
# fill missing value
df = df.fillna(0)

df
```

```python
# The goal for this part is to find the correlation of feature and result
# 但是全部特徵都考慮的話, 準確率測試起來會最高
import seaborn as sns
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True)
```

#### 先分開input跟output
#### 刪掉離群值(outlier)
#### 標準化資料
#### 用polynomial linear regression (比一般一次方的線性回歸效果好一點)

```python
# Define X (input features) and y (target variable)
X = df[['datetime', 'season', 'holiday', 'workingday', 'weather', 'temp', 'atemp', 'humidity', 'windspeed']]
y = df['count']

# Split the data into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 1. use IQR method to identify the outlier.

# merge two dataframe first
combine = X_train.copy()
combine['result'] = y_train

# Define a function to identify and remove outliers
col = ['datetime', 'season', 'holiday', 'workingday', 'weather', 'temp', 'atemp', 'humidity', 'windspeed']
def remove_outliers(df, columns, threshold=1.5):
    outliers = pd.DataFrame()
    for col in columns:
        # quantile to select the 第幾分位數
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        column_outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outliers = pd.concat([outliers, column_outliers], axis=0)
    return df[~df.index.isin(outliers.index)]

# pick out outlier
combine = remove_outliers(combine, col)
# Split into the two dataframe
X_train = combine.drop(columns=['result'])
y_train = combine['result']

# 2. standardize
scaler = StandardScaler()
# 將X_train 的 feature標準化
data_to_standardize = X_train[['datetime', 'season', 'holiday', 'workingday', 'weather', 'temp', 'atemp', 'humidity', 'windspeed']] 
scaler.fit(data_to_standardize)
standardized_data = scaler.transform(data_to_standardize)
X_train[['datetime', 'season', 'holiday', 'workingday', 'weather', 'temp', 'atemp', 'humidity', 'windspeed']] = standardized_data

# 3. polynomial linear regression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
# Create a Polynomial Regression model
degree = 2
model = make_pipeline(PolynomialFeatures(degree, include_bias=False, interaction_only = True), LinearRegression())

# Fit the model on the training data
model.fit(X_train, y_train)
```

#### 測試MSE 跟 r2_score

```python
# 將X_test 的 feature標準化
data_to_standardize = X_test[['datetime', 'season', 'holiday', 'workingday', 'weather', 'temp', 'atemp', 'humidity', 'windspeed']] 
scaler.fit(data_to_standardize)
standardized_data = scaler.transform(data_to_standardize)
X_test[['datetime', 'season', 'holiday', 'workingday', 'weather', 'temp', 'atemp', 'humidity', 'windspeed']] = standardized_data

# Test Mean Square Error (The loss function)
y_pred = model.predict(X_test)
from sklearn.metrics import mean_squared_error
print("MSE:" , mean_squared_error(y_test,y_pred))

# measures the goodness of fit of the model.
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print("r2_score: ", r2)
```

#### 測試正規化(Regurlarization)
#### 但是發現他們只有 好一點點

```python
# Test Regularization
# Create a polynomial regression model with Ridge and Lasso regularization
degree = 2  # Degree of the polynomial features
alpha = 30  # Ridge regularization parameter
L_alpha = 0.1
Ridge_reg = make_pipeline(PolynomialFeatures(degree), Ridge(alpha=alpha))
Lasso_reg = make_pipeline(PolynomialFeatures(degree), Lasso(alpha=L_alpha))
Ridge_reg.fit(X_train, y_train)
Lasso_reg.fit(X_train, y_train)

print("Ridge regression:")
# Predict with the model
y_pred = Ridge_reg.predict(X_test)

# Calculate mean squared error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.4f}")
# Calculate the r2_score
r2 = r2_score(y_test, y_pred)
print("r2_score: ", r2)

print()

print("Lasso regression:")
# Predict with the model
y_pred = Lasso_reg.predict(X_test)

# Calculate mean squared error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.4f}")
# Calculate the r2_score
r2 = r2_score(y_test, y_pred)
print("r2_score: ", r2)
```

#### 用cross validation 再次檢驗
#### 準確率落在 三成 左右

```python
# cross validation folding
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# cross validation score
cross_val_scores = cross_val_score(model, X, y, cv=kf)

# output score
for i, score in enumerate(cross_val_scores):
    print(f"Fold {i+1}: {score:.4f}")

# average
print(f"Average Cross-Validation Score: {cross_val_scores.mean():.4f}")
```

```python
# 知道模型的選擇出的參數
# coefficient in linear regression
coef = model.named_steps['linearregression'].coef_
print("coeffiecient:" , coef)
# 獲取截距
bias = model.named_steps['linearregression'].intercept_
print("bias:" , bias)
print()
```
